```python
#Iimpprt package
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn as sk

#Membaca data dari file CSV dan menyimpannya ke dalam objek df
df = pd.read_csv('heartpredict.csv')
print(df.head()) #5 baris pertama
```

```python
#Menampilkan statistik deskriptif dari objek df
df.describe()
```

```python
#Cek mising pada setiap kolom, jika 0 menandakan aman dari mising
df.isna().sum()
```

```python
#Visualisasi Grafik dengan Menampilkan Scatter
#Membuat scatter plot untuk variabel 'age' dan 'thalach' (dengan warna yang berbeda)
sns.scatterplot(data=df, x='age', y='thalach', hue='sex')

#Label pada X dan Y
plt.xlabel('Age')
plt.ylabel('Thalach')

plt.title('Scatter Plot Age vs Thalach')
plt.show()
```

Pada visualisasi scatter plot Age Vs Talach menunjukkan bahwa umur dan detak jantung maksimum pada saat seseorang melakukan tes latihan jantung.

Dari hasil gambar scatter tersebut dapat kita simpulkan bahwa semakin tua usia seseorang, maka semakin rendah maksimum detak jantungnya. Seperti pada titik yang berkumpul di bagian bawah kiri yang menunjukkan masih ada beberapa pasien dimana usia yang terpaut tinggi dan detak jantung maksimum yang rendah. Walaupun memang scatter yang dihasilkan belum maksimal sempurna karna masih ada pengecualian di beberapa titik yang tersebar.

```python
#Visualisasi Grafik dengan Menampilkan Matrix Korelasi
corr_matrix = df.corr()

#Show heatmap dari matrix korelasi
sns.heatmap(corr_matrix, annot=True)

plt.title('Matriks Korelasi')
plt.show()
```

Matriks korelasi pada dataset heartpredict menunjukkan korelasi antara setiap pasang variabel dalam bentuk angka dimana semakin tinggi nilai korelasi, maka semakin kuat hubungan antara dua variabel tersebut.

Seperti pada gambar diatas dimana korelasi terkuat antara dua variabel ditunjukkan oleh warna terang seperti warna putih atau kuning, sedangkan korelasi terlemah ditunjukkan oleh warna yang lebih gelap seperti warna biru atau ungu.
